```python
import pandas as pd
sp500_tickers = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0].Symbol.to_list()
sp500_tickers = [ticker for ticker in sp500_tickers if '.B' not in ticker]
```

```python
sp500_tickers
```

```python
# define dates
start_date = '2000-01-01'
end_date = '2024-11-01'
historical_prices = yf.download(sp500_tickers, start=start_date, end=end_date)
```

```python
prices = historical_prices.loc[:, historical_prices.columns.get_level_values(0) == 'Adj Close']
```

```python
prices.columns = prices.columns.droplevel(0)
```

```python
prices.head()
```

```python
MIN_REQUIRED_OBS_PER_TICKER = 100

ticker_counts = prices.count()
ticker_counts
```

```python
valid_ticker_mask = ticker_counts[ticker_counts >= MIN_REQUIRED_OBS_PER_TICKER].index
valid_ticker_mask
```

```python
prices = prices[valid_ticker_mask]
```
